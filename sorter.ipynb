```python
import pandas as pd
```

```python
df = pd.read_csv('PokemonStats.csv', encoding='cp949')
df
```

```python
# Function to split types
def split_types(type_str):
    upper_case_indices = [i for i, c in enumerate(type_str) if c.isupper()]
    if len(upper_case_indices) < 2:
        return type_str, None
    split_index = upper_case_indices[1]
    return type_str[:split_index], type_str[split_index:]

# Applying the function and creating new columns
df['Type1'], df['Type2'] = zip(*df['Type'].apply(split_types))

# Dropping the original 'Type' column
df.drop(columns='Type', inplace=True)

# Rearranging columns to have Type1 and Type2 at the position of the original Type
column_order = ['No', 'Pokemon', 'Type1', 'Type2', 'Abilities', 'HP', 'Atk', 'Def', 'SpA', 'SpD', 'Spe', 'BST']
df = df[column_order]

df
```

```python
df.to_csv('PokemonStats2.csv', index=False, encoding='cp949')
```

```python
pokemons_df = pd.read_csv('pokemons.csv', encoding='UTF-8')
kitakami_df = pd.read_csv('kitakami.csv', encoding='UTF-8')

# Convert names in both dataframes to lowercase to ensure matching is case-insensitive
pokemons_df['name'] = pokemons_df['name'].str.lower()
kitakami_df['name'] = kitakami_df['name'].str.lower()

# Merge the required columns based on the 'name' column from both dataframes
merged_df = kitakami_df.merge(
    pokemons_df[['name', 'hp', 'atk', 'def', 'spatk', 'spdef', 'speed', 'abilities']], 
    on='name', 
    how='left'
)

# Display the first few rows of the merged dataframe to verify the merge
merged_df
```

```python
# Removing any extra spaces and the subsequent text from the 'abilities' column
merged_df['abilities'] = merged_df['abilities'].str.split().str[0]

# Display the updated dataframe
merged_df
```

```python
# Replacing hyphens with spaces in the 'abilities' column
merged_df['abilities'] = merged_df['abilities'].str.replace('-', ' ')

# Display the updated dataframe
merged_df
```

```python
merged_df.to_csv('kitakami_stats.csv', index=False)
```

```python
kitakami_statsus_df = pd.read_csv('kitakami_statsus.csv', encoding='euc-kr')
pokemon_stats_df = pd.read_csv('PokemonStats.csv', encoding='euc-kr')

pokemon_stats_df
```

```python
kitakami_statsus_df
```

```python
# Find names in 'kitakami_stats_df' that are not in 'pokemon_stats_df'
unique_names = kitakami_statsus_df[~kitakami_statsus_df['Name'].isin(pokemon_stats_df['Name'])]

# Assign new 'No' starting from 401
unique_names['No'] = range(401, 401 + len(unique_names))

# Append the unique rows to 'pokemon_stats_df'
updated_pokemon_stats_df = pd.concat([pokemon_stats_df, unique_names], ignore_index=True)
updated_pokemon_stats_df
```

```python
updated_pokemon_stats_df.to_csv('PokemonStats2.csv', index=False, encoding='euc-kr')
```
